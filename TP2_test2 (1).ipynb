```python
# Define the insertion sort algorithm
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
# Extract the column to be sorted
est_hhc = df['EST_HHC'].dropna().astype(int).tolist()  
```

```python
# Sort the column using insertion sort
sorted_est_hhc = insertion_sort(est_hhc)
```

```python
# Replace the sorted values back in the DataFrame
df_sorted = df.copy()
df_sorted.loc[df['EST_HHC'].notna(), 'EST_HHC'] = sorted_est_hhc
```

Linear search and binary search are two fundamental algorithms used to search for elements in a list. Here's an explanation of their performance:

Linear Search:

-**Performance**: O(n) time complexity.
How it works: Linear search iterates through each element in the list until it finds the target element or reaches the end of the list.
Pros: Simple and can be used on any list, sorted or unsorted.
Cons: Inefficient for large lists since it may need to check every element.
Binary Search:

Performance: O(logn) time complexity.

How it works: Binary search repeatedly divides the search interval in half. It starts with the middle element of a sorted list and eliminates half of the remaining elements based on comparisons.

Pros: Much faster than linear search for large, sorted lists.
Cons: Requires the list to be sorted.

In terms of performance, binary search performs faster than linear search because it reduces the search space by half with each step, making it logarithmic in nature. However, binary search can only be used on sorted lists, whereas linear search can be used on both sorted and unsorted lists.

#### Time Complexity

Insertion sort works by iteratively taking an element from the unsorted portion of the list and inserting it into its correct position in the sorted portion. Hereâ€™s the step-by-step breakdown of the time complexity:

1. **Best Case**:
   - When the list is already sorted, each element only needs to be compared once with the previous element.
   - Time complexity: \(O(n)\).

2. **Average Case**:
   - On average, each element will be compared with half of the elements in the sorted portion.
   - Time complexity: \(O(n^2)\).

3. **Worst Case**:
   - When the list is sorted in reverse order, each element must be compared with every other element in the sorted portion.
   - Time complexity: \(O(n^2)\).

In summary, the time complexity of the insertion sort algorithm is \(O(n^2)\) in the average and worst cases, and \(O(n)\) in the best case.

#### Space Complexity

Insertion sort is an in-place sorting algorithm, which means it does not require additional storage space for another list or array. Hereâ€™s the breakdown of the space complexity:

1. **Auxiliary Space**:
   - Insertion sort uses a constant amount of extra space.
   - Space complexity: \(O(1)\).

In summary, the space complexity of the insertion sort algorithm is \(O(1)\).

#### Efficiency

Insertion sort is efficient for small datasets or nearly sorted datasets due to its \(O(n)\) best-case time complexity. However, it becomes inefficient for larger datasets due to its \(O(n^2)\) average and worst-case time complexity.

The simplicity of insertion sort and its in-place sorting nature make it a good choice for scenarios where additional memory usage needs to be minimized, and the dataset size is manageable.
